## Custom Transformers

```python
from sklearn.base import BaseEstimator, TransformerMixin

class CustomAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bmi_category=True):
        self.add_bmi_category = add_bmi_category
        
    def fit(self, X, y=None):
        # Nothing to fit in this case, just return self
        return self
    
    def transform(self, X):
        # Extract relevant columns by index
        bmi_ix = 8
        
        # Create BMI category: 0 = Underweight, 1 = Normal, 2 = Overweight, 3 = Obese
        bmi_values = X[:, bmi_ix]
        bmi_category = np.digitize(bmi_values, bins=[18.5, 24.9, 29.9, 40])  # BMI categorization
        
        # Prepare the new feature list
        new_features = []
        
        # Optionally add BMI category feature
        if self.add_bmi_category:
            new_features.append(bmi_category)
        
        # Combine the new features with the original dataset
        return np.c_[X, *new_features]
```

```python
# Initialize the custom transformer
attr_adder = CustomAttributesAdder(add_bmi_category=True)

# Apply it to your dataset
transformed_data = attr_adder.transform(df.values)

# Convert back to DataFrame (optional)
df = pd.DataFrame(transformed_data, columns=df.columns.tolist() + ['bmi_category'])
df
```
